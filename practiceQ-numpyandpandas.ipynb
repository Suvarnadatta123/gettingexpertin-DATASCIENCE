```python
import numpy as np
import pandas as pd

# NumPy (no column names)
arr = np.array([[1, 85], [2, 90]])
print(arr)

# Pandas (with labels)
df = pd.DataFrame(arr, columns=["ID", "Marks"])
print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Marks": [80, 90, 85]
}, index=["x", "y", "z"])

# loc (label)
print(df.loc["x"])

# iloc (position)
print(df.iloc[0])
```

```python
# handleing missing values
# ans : Pandas provides functions like fillna() and dropna()
import pandas as pd

df = pd.DataFrame({
    "A": [1, None, 3],
    "B": [4, 5, None]
})

# Fill missing with mean
df_filled = df.fillna(df.mean())

# Drop missing
df_dropped = df.dropna()

print(df_filled)
print(df_dropped)
```

```python
# broadcasting in NumPy
# ans : Broadcasting allows operations between arrays of different shapes.
import numpy as np

arr = np.array([1, 2, 3])
result = arr + 5

print(result)   # [6 7 8]
```

```python
# GroupBy in Pandas
# ans : Used to group data and apply aggregation functions.
import pandas as pd

df = pd.DataFrame({
    "Dept": ["IT", "HR", "IT"],
    "Salary": [50000, 40000, 60000]
})

result = df.groupby("Dept")["Salary"].mean()
print(result)
```

```python
# Merge vs Concat
# ans : merge() → combine on keys (like SQL join)   concat() → stack data
import pandas as pd

df1 = pd.DataFrame({"ID": [1, 2], "Name": ["A", "B"]})
df2 = pd.DataFrame({"ID": [1, 2], "Marks": [80, 90]})

# Merge
merged = pd.merge(df1, df2, on="ID")

# Concat
concat = pd.concat([df1, df2])

print(merged)
print(concat)
```

```python
# Apply Function
# Ans: Used to apply custom logic to columns.
import pandas as pd

df = pd.DataFrame({"Marks": [30, 50, 80]})

df["Result"] = df["Marks"].apply(lambda x: "Pass" if x > 40 else "Fail")

print(df)
```

```python
# Remove Duplicates
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "A"],
    "Marks": [80, 90, 80]
})

df_unique = df.drop_duplicates()
print(df_unique)
```

```python
# Top N Values
import pandas as pd

df = pd.DataFrame({"Marks": [50, 90, 80, 95]})

top3 = df.nlargest(3, "Marks")
print(top3)
```

```python
# Find highest salary per department.
import pandas as pd

df = pd.DataFrame({
    "Dept": ["IT", "HR", "IT"],
    "Salary": [50000, 40000, 60000]
})

result = df.groupby("Dept")["Salary"].max()
print(result)
```
