```python
# label based indexing (.loc)
import pandas as pd

data = {'Name': ['A', 'B', 'C'], 'Marks': [90, 85, 88]}
df = pd.DataFrame(data)

df.loc[0, 'Name']
```

```python
# position based indexing (.iloc)
df.iloc[0, 0]
```

```python
df.loc[0] # First row (by label)
df.iloc[0] # First row (by position)
```

```python
df.loc[0:2, ["Name", "Marks"]] # Rows 0 to 2, selected columns
df.iloc[0:2, 0:2] # Rows and cols by index position
```

👉 .loc[] → “Give data by name”
👉 .iloc[] → “Give data by position”

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Age": [20, None, 25],
    "Gender": ["Male", "Female", "Male"]
}

df = pd.DataFrame(data)

# Transformation steps
df["Age"] = df["Age"].fillna(df["Age"].mean())  # handle missing
df["Gender"] = df["Gender"].map({"Male":0, "Female":1})  # encode
df["Age"] = df["Age"].astype(int)  # type conversion

print(df)
```
