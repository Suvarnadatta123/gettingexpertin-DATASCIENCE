```python
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Age": [20, 25, 30]
}

df = pd.DataFrame(data)
print(df)
#Industry use: APIs, JSON → converted to DataFrame
```

```python
#From List of Lists
data = [
    ["A", 20],
    ["B", 25],
    ["C", 30]
]

df = pd.DataFrame(data, columns=["Name", "Age"])
df
```

```python
#From Dictionary of Lists

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}
 
df = pd.DataFrame(data)
df
#Each key becomes a column, and each list is the column data.
```

```python
#From NumPy Arrays
import numpy as np
 
arr = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(arr, columns=["A", "B"])
df

#Make sure to provide column names!
```

```python
#Selection (Accessing Data)
import pandas as pd

df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "Age": [20, 25, 30],
    "Marks": [80, 90, 85]
})
df["Name"]
df[["Name", "Marks"]]
```

```python
#Selecting Rows
#✔️ Using .iloc (index-based)
df.iloc[0]        # first row
df.iloc[0:2]      # slice rows

#✔️ Using .loc (label-based)
df.loc[0]
```
