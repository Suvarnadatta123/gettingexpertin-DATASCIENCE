```python
import json

data = {"name": "Raj", "age": 20}

json_data = json.dumps(data)   #json.dumps() (Python → JSON string)
print(json_data)
```

```python
import json

json_data = '{"name": "Raj", "age": 20}'

data = json.loads(json_data) #json.dump() (Python → JSON file)
print(data["name"])
```

```python
import json

data = {"name": "Priya", "age": 25}

with open("data.json", "w") as file:
    json.dump(data, file)  #json.dump() (Python → JSON file)
```

```python
import json

with open("data.json", "r") as file:
    data = json.load(file) #json.load() (JSON file → Python)

print(data)
```
