```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)          # create plot
plt.title("Sample Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()              # display graph
```

```python
# Sales Trend Analysis (Line Plot)
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [200, 250, 300, 280, 350]

plt.plot(months, sales, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
# Product Comparison (Bar Chart)
products = ["Laptop", "Mobile", "Tablet", "Headphones"]
sales = [150, 300, 100, 200]

plt.bar(products, sales)
plt.title("Product Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
# Customer Age Distribution (Histogram)
ages = [18, 22, 25, 30, 35, 40, 22, 25, 30, 28, 35, 40]

plt.hist(ages, bins=5)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Study Hours vs Marks (Scatter Plot)
hours = [1, 2, 3, 4, 5]
marks = [40, 50, 60, 70, 80]

plt.scatter(hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.show()
```

```python
# Market Share Analysis (Pie Chart)
brands = ["Brand A", "Brand B", "Brand C", "Brand D"]
share = [40, 30, 20, 10]

plt.pie(share, labels=brands, autopct='%1.1f%%')
plt.title("Market Share")
plt.show()
```
